```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## load boston house price prediction 

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
# the below shows the details under the dataset. 'data' is the actual data. feature_names are the name of the columns of data.
#target is the dependant variable which is the price of the houses. DESCR gives the details of the dataset. We can access all 
#details under the keys using 'dot' operator
boston.keys()
```

```python
# see the desfcription of dataset
print(boston.DESCR)
```

```python
#let us check the data. the data is shown in terms of arrays.
print(boston.data)
```

```python
#convert the data in to pandas dataframe
df = pd.DataFrame(boston.data,columns=boston.feature_names)
df
```

```python
# assigning the target feature
df['Price'] = boston.target 
```

```python
# checki8ng the first five rows 
df.head(5)
```

```python
# check information about dastaset
df.info()
```

```python
# summarize the stats of dataset
df.describe()
```

```python
# check for missing values
df.isnull().sum()
```

### Explorqatory Data Analysis

```python
# check for correlation
df.corr()
```

```python
# plot the correlation of the features with each other and with the target 
plt.figure(figsize=(12,7))
sns.heatmap(df.corr(),annot = True)
```

```python
sns.pairplot(df.corr())
```

```python
# dependent and independent features 
X = df.drop(columns='Price')
y = df['Price']
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# standardise the datset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

## Model Training

```python
from sklearn.linear_model import LinearRegression 
```

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
# print the coefficient
regression.coef_
```

```python
# print the intercept
regression.intercept_
```

```python
regression.get_params()
```

```python
# predictiong with the test data
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

### Assumptions

```python
# plot scatterplot for prediction
plt.scatter(y_test,reg_pred)
```

```python
# redisuals 
residuals = y_test-reg_pred 
```

```python
# plotting the residuals
sns.displot(residuals,kind='kde')
```

```python
# calculating mean square and mean absolute eroor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mse = mean_squared_error(y_test,reg_pred)
mae = mean_absolute_error(y_test,reg_pred)
```

```python
print('mse-',mse,'\nmae-',mae)
print('rmse-',np.sqrt(mse))
```

```python
# r2 square 
from sklearn.metrics import r2_score
r2s = r2_score(y_test,reg_pred)
print("r2_score-",r2s)
```
